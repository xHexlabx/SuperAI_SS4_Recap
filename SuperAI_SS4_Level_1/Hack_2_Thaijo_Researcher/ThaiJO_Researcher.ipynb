## Filter Function

```python
# has english ?
def has_english_in_parens(text):
  return bool(re.search(r"\([\w\s]+\)", text))
```

```python
# case : "บัณฑิตา (Banthita) หล่อนิมิตดี (Lornimiddee)"
pattern = r'([^\(]+)\(([^\)]+)\)'
def thai_eng_double_bracket(text):
    matches = re.findall(pattern, text)
    if matches and len(matches)==2:
        thai_name = matches[0][0].strip() + ' ' + matches[1][0].strip()
        eng_name = matches[0][1].strip() + ' ' + matches[1][1].strip()
        text = [thai_name, eng_name]
    elif matches and len(matches)==1:
        thai_name = matches[0][0].strip()
        eng_name = matches[0][1].strip()
        if len(re.split(" ",eng_name))>=2 or has_english_in_parens(text):
            text = [thai_name, eng_name]
    return text
```

```python
# prefix ชาย
def remove_nai(entries):
    return [re.sub(r"(Editorial Team|Author JPHD|พระครูสังฆรักษ์|บรรณาธิการ|พระเทพ|พระมหาอุดร |พระมหา|พระครู|\]| \[| \-|ผศ\.|อ\.|ม\.ร\.ว\.|ว่าที่ ร\.ต\. |ว่าที่ พ\.ต\. |ผศ\.ดร\.|ดร\.|นางสาว|^Dr\. |^Dr\.|^รศ\.|^รศ\.ดร\. |^อาจารย์ ดร\.|^อาจารย์ |^ว่าที่ร้อยตรี|^ว่าที่ รต |^ว่าที่ พต |\*$|^นาย|^นาง|\"|, |ผู้ช่วยศาสตราจารย์ |รองศาสตราจารย์ |^\(|^\) |สำนักส่งเสริมวิชาการและงานทะเบียน มรภวไลยอลงกรณ์ ในพระบรมราชูปถัมภ์|บทบรรณาธิการ \(Editorial\|Journal of Thai Traditional Alternative Medicine|พัฒนาตำราและวารสาร วิทยาลัยดุสิตธานี Dusit Thani College|คณะนักศึกษาหลักสูตรป้องกันราชอาณาจักร รุ่นที่ 57)", "", entry) for entry in entries]
```

```python
# พระ aka แซลมอน
def remove_pha(entries):
    return [re.sub(r"(^สังฆรักษ์|เจ้าหน้าที่สาธารณสุขโรงพยาบาลส่งเสริมสุข| M\.hum| Sahufi L\.|ร\.ต\.| ประพันธ์วงศ์$|^อุดร | Phithakphongphan$|พระครูปลัดบำรุง |พระมหา|พระครู)","", entry) for entry in entries]
```

```python
# MA
def remove_isolated_ma(text):
  pattern = r"\bMA\b"
  return [re.sub(pattern, "", entry) for entry in text]
```

```python
# นัฏยา ประดับการ Nattaya Pradubkarn
thai_pattern = r"([\u0E00-\u0E7F]+)"

def separate_thai_eng(text):
    pattern = r"([\u0E00-\u0E7F]+ [a-zA-Z0-9]+)"
    out = []

    for entry in text:
        if re.search(pattern, entry):
            matches = re.split(r"( )", entry)
            if len(matches) == 7:
                thai_name1, thai_name2, eng_name1, eng_name2 = matches[0], matches[2], matches[4], matches[6]
                thai_name = f"{thai_name1} {thai_name2}"
                eng_name = f"{eng_name1} {eng_name2}"
                out.extend([thai_name, eng_name])
            else:
                out.append(entry)
        else:
            out.append(entry)
    return out
```

```python
def fix_name_case(name):
    execption = ["E.B.","A.B.","Phokhwang-Just","B.M.","F.K.","Yow-Mow","C.A","U.N.","Sitti-Amorn",
    "O’Connor","Sae-Jong","SU","Chi-Wei","Chunthorng-Orn","Ya-Liang","KongpanJittima","Chai-adisaksopha","Sang-ubon",
    "O-Tur","M.N.","Swangchan-Uthai","Na-nakorn","of","Thong-Aram","Sudasna-na-Ayudthya","bin","San-Um","Lee-Shuan",
    "Settheetham-Ishida","Okafor-Terver","McLoughlin","Ua-Kit","MuangKaow","Yun-Bae-Kim","O'Connell","Im-Aim","NumChai",
    "T.C.","Le-Thanh","N.K.","A.R.M.","G.R.","S.S.","V.S.","LOUG-UI1","Eam-Urai","A.A.","K.O.","E.O.","S.A.",
    "Ratana-Olarn","Jit-Aree","Jean-Noël","G.J","EI-sayed","El-Aziz","Sae-Lee","McDaniel","O-Prasertsawat","Sitthi-Amorn",
    "Jan-Frits","Uan-On","J.S.","I-Ming","McDonough","Riam-Amatakun","R.S.","M.Q.","N.J.","Amir-Ali"]
    fixed_parts = [part.capitalize() if part not in execption else part for part in name.split()]
    return ' '.join(fixed_parts)
```

```python
# MA
def addon_case(text):

  return [re.sub(pattern, "", entry) for entry in text]
```

## Main Function (Get Text)

```python
# case : ยุพาวรรณ ดุลยสุจริต Yupawan  Dulayasutchari
elimination_list = [' Dr.'," ","",",","(",")","และคณะ",'et.al','et al', '. .','- -'
                    ,'-- --',"Authors :","B. Pharm.","M. Pharm.","TSME Thailand","Editorial","บทบรรณาธิการ"]
common_prefix = ['M.R.','M.D.','MA','CFA','Jr.', 'ดร.', 'Ph.D']
ans = pd.read_csv("/content/drive/MyDrive/Submission3.csv")
def get_names(text):
    if pd.notnull(text):
        res = re.split(r"(,)",text)
        res = [element.strip() for element in res if "Vol.12" not in element]
        res = ",".join(res)
        res = re.sub("(TSME Thailand|Cover SNRUJST|subcommittee on the preparation of monofraphs of selected thai materia medica|subcommittee on the preparation of Monographs of Selected Thai Materia Medica| และคณะ|พระครูใบฎีกา)","",res)
        res = re.sub(" and Other","",res)
        res = re.sub(" Authors :","",res)
        res = re.sub("   "," ",res)
        res = re.sub("  "," ",res)
        res = re.sub("( 1 | i )"," ",res)
        res = re.sub(" ๋j"," J",res)
        res = re.sub("O’CONNELL","O'Connell",res)
        res = re.sub("LE-THANH","Le-Thanh",res)
        res = re.sub("DwiJoko","Dwi Joko Suroso",res)
        res= re.sub("RIAM-AMATAKUN","Riam-Amatakun",res)
        res = re.sub("SAE-JOO","Sae-Joo",res)
        res = re.sub(r"(^Ms\.|^Mr\.|^M\.L\.|^์|^ื|คณะนักศึกษาหลักสูตรป้องกันราชอาณาจักร รุ่นที่ 57|เกี่ยวกับวารสาร ABOUT THE JOURNAL|^พระมหา|ู^พระครู|Ven\. Dr\. |Assoc\.Prof\.Dr\.|Asst\.Prof\.Dr\.|ดร\. |Lect\. Dr\. |Assist\. Prof\. Dr\.|Assoc\. Prof\. Dr\. )","",res)
        res = re.split(r"(,|and Other|Authors :| \[| Dr\.)", res)
        res = [re.sub(r"( \[|\])","",entry) for entry in res]
    else:
        res = []

    res = [element.strip() for element in res if element not in elimination_list]

    bracket_managed = []
    for idx,names in enumerate(res):
        process = thai_eng_double_bracket(names)
        if isinstance(process, list):
            bracket_managed.extend(thai_eng_double_bracket(names))
        else:
            bracket_managed.append(thai_eng_double_bracket(names))

    for idx,names in enumerate(bracket_managed):
        bracket_managed[idx] = re.sub("(รศ\.ดร\. |อ\.ดร\.|ผศ\.ดร\.|M\.R\.|M\.D\.|CFA|Jr\.|Ph\.D)","",names)

    special_case_managed = []
    for idx,names in enumerate(bracket_managed):
        if names == "กัลยา บุญหล้า และศศิธร โกฎสืบ":
            special_case_managed.extend(["กัลยา บุญหล้า","ศศิธร โกฎสืบ"])
        else:
            special_case_managed.append(names)
#  Fillter Case
    special_case_managed = remove_nai(special_case_managed)
    special_case_managed = remove_isolated_ma(special_case_managed)
    special_case_managed = separate_thai_eng(special_case_managed)
    special_case_managed = remove_pha(special_case_managed)
    special_case_managed = [element.strip() for element in special_case_managed if element not in elimination_list]
    special_case_managed = [fix_name_case(name) for name in special_case_managed]
    return special_case_managed
ans.to_csv("Submission3.csv" , index=False)
```
