# SuperAI Season 4 - Level Individual Hackathon - Hack_Car_Prediction

## Explore Data

```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('./datasets/train.csv' , index_col = 'Id')
train_target = pd.read_csv('./datasets/train_target.csv' , index_col = 'Id')
test_df = pd.read_csv('./datasets/test.csv', index_col = 'Id')
```

```python
train_df
```

```python
def apply_features(row) : 
    
    for col in train_df.columns :
        
        if row[col] == 888 or row[col] == 999 or row[col] == '888' or row[col] == '999':
            
            row[col] = np.nan
    
    return row
```

```python
train_df = train_df.apply(apply_features , axis = 1)
test_df = test_df.apply(apply_features , axis = 1)
```

```python
train_df
```

```python
test_df
```

```python
columns = ['Q1_OccCode' , 'Q2.1_OccCode' , 'Q2.2_OccCode' , 'Q3.3Reason' , 'Q4_7' , 'Q9_9' , 'Facebook' , 'Instragram' , 'Tiktok' , 'Youtube' , 'Twitter' , 'OtherApp' , 'OtherAppSpec' , 'NoSocial']
```

```python
train_df = train_df.drop(columns=columns)
test_df  = test_df.drop(columns=columns)
```

## AutoGluon

```python
from autogluon.tabular import TabularPredictor
```

```python
label = 'Predicted'
```

```python
predictor = TabularPredictor(label = label , eval_metric='accuracy').fit(train_df.join(train_target))
```

```python
predictor.refit_full()
```

```python
predictor.fit_pseudolabel(test_df)
```

```python
prediction = predictor.predict(test_df)
```

```python
prediction
```

```python
features_importance = predictor.feature_importance(train_df.join(train_target))
features_importance
```

```python
features_importance.to_csv('./features/features.csv')
```

```python
submission_path = './submissions'
prediction.to_csv(f'{submission_path}/submission_refit.csv')
```
