```python
def add_features (row) :
    
    year , month , day = row['time'].split(' ')[0].split('-')
    hour = row['time'].split(' ')[1].split(':')[0]
    
    year , month , day , hour = int (year) , int (month) , int (day) , int (hour) 
    
    row['month'] = month
    row['day'] = day
    row['hour'] = hour 
    
    return row
```

```python
train_df = train_df.apply(add_features , axis = 1).drop(columns=['time' , 'rainfall' , 'wind_speed' , 'wind_direct'])
test_df = test_df.apply(add_features , axis = 1).drop(columns=['time' , 'rainfall' , 'wind_speed' , 'wind_direct'])
```

```python
train_df = train_df.fillna(train_df.mean())
test_df = test_df.fillna(test_df.mean())
```

## Catboost / XGBoost

```python
from catboost import CatBoostRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split , cross_val_score, KFold
from sklearn.metrics import mean_absolute_error
```

```python
catboost_model = CatBoostRegressor(n_estimators = 1000 , learning_rate = 0.2, depth = 12 , random_seed = 42 , loss_function = 'MAE')
xgb_model = XGBRegressor(n_estimators= 1000 ,objective = 'reg:absoluteerror' , learning_rate = 0.2 , eval_metric = 'mae' , max_depth = 12 )
```

```python
cat_scale = 0.6
xgb_scale = 1 - cat_scale
```

```python
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)

cv_scores = []

train_data = train_df.drop(columns = 'temp') 
label_data = train_df['temp']

for train_index, val_index in kf.split(train_data):
    
    x_train, x_val = train_data.iloc[train_index], train_data.iloc[val_index]
    y_train, y_val = label_data.iloc[train_index], label_data.iloc[val_index]
    
    catboost_model.fit(x_train, y_train , verbose = 500)
    xgb_model.fit(x_train, y_train)
    
    val_predictions = cat_scale * catboost_model.predict(x_val) + xgb_scale * xgb_model.predict(x_val)
    
    mae = mean_absolute_error(y_val, val_predictions)
    cv_scores.append(mae)
    
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", np.mean(cv_scores))
```

```python
submission = pd.read_csv('./submissions/sample_submissions.csv' , index_col='id')
```
